```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mlp
%matplotlib inline
%config InlineBackend.figure_format="retina"
```

```python
def circle(a,b,r):
    fig,ax =plt.subplots(figsize=(5,5))
    t=np.linspace(0,np.pi*2,90)
    x = a+r *np.cos(t)
    y = b+r *np.sin(t)
    ax.axvline(a,color="gray",linewidth=.5,linestyle='--')
    ax.axhline(b,color="gray",linewidth=.5,linestyle='--')
    ax.plot(a,b,color="r",marker="o")
    rect=mlp.patches.Rectangle((-.5,-.5),1,1,fill=False)
    ax.add_patch(rect)
    ax.axis("off")
    ax.plot(x,y)
    
#     def circle(a, b, r):
#     fig, ax = plt.subplots(figsize=(5,5))
#     t = np.linspace(0, np.pi * 2, 90)
#     x = a + r * np.cos(t)
#     y = b + r * np.sin(t)

#     ax.axvline(a, color='gray', linewidth=.5, linestyle='--')
#     ax.axhline(b, color='gray', linewidth=.5, linestyle='--')
#     ax.plot(a, b, color='r', marker='o') # center point
    
#     rect=mpl.patches.Rectangle((-.5, -.5), 1, 1, fill=False)
#     ax.add_patch(rect)
#     ax.axis('off')
    
#     ax.plot(x, y)
```

```python
circle(0,0,.5)
```

```python
n = 1000000
x=np.random.uniform(-.5, .5, n)
y=np.random.uniform(-.5, .5, n)
a=x**2 + y**2
plt.figure(figsize=(5,5))
plt.scatter(x[a < .25], y[a < .25], s=.1)
plt.scatter(x[a >= .25], y[a > .25], color='r', s=.1)
pi_est=(a < .25).sum() / n * 4
plt.title(f'n = {n:,}, estimated pi = {pi_est}, delta =  {np.abs(np.pi - pi_est)}')
plt.axis('off');
```
